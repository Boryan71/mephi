```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
import seaborn as sns
```

Задание №1. Проведите корреляционный анализ данных из файла 'ex1.csv' по Пирсону

```python
# считывание файла
df1 = pd.read_csv('./ex1.csv') 
df1.head()
```

```python
# КРИТЕРИЙ 1

# проверка, что все элементы - численного типа
print('Типы данных:')
print(df1.dtypes.to_string())

# построение гистограмм распределения
plt.hist(df1['x'], bins=20)
plt.title('Гистограмма распределения x')
plt.xlabel('x')
plt.ylabel('Частота')
plt.show()

plt.hist(df1['y'], bins=20)
plt.title('Гистограмма распределения y')
plt.xlabel('y')
plt.ylabel('Частота')
plt.show()

# построение "ящика с усами" для изучения выбросов
sns.boxplot(df1['x'])
plt.title('"Ящик с усами" для x')
plt.show()

sns.boxplot(df1['y'])
plt.title('"Ящик с усами" для y')
plt.show()

# построение диаграммы рассеяния
plt.scatter(df1['x'], df1['y'])
plt.title('Диаграмма рассеяния x и y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# вывод о целесообразности использования корреляции Пирсона с обоснованием на основе предшествующих графиков
print("""На основе построенных графиков, можем сказать, что существует линейная зависимость между x и y,
следовательно использование корреляции Пирсона целесообразно.""")
```

```python
# КРИТЕРИЙ 2
from scipy.stats import pearsonr
from scipy.stats import t

print('Вычисленные значения:')
# расчет коэффициента Пирсона
corr, p_value = pearsonr(df1['x'], df1['y'])
print(f'Коэффициент Пирсона: {corr:.3f}')
print(f'p-value: {p_value}')

# расчет t-критерия
t_stat = corr * ( (len(df1)-2)/(1-corr**2) )**0.5
print(f't-статистика: {t_stat:.3f}')

# t-критическое при alpha = 0.05
alpha = 0.05
t_crit = t.ppf(1-alpha/2, len(df1)-2)
print(f't-критическое: ±{t_crit:.3f}')
print()

# вывод о степени корреляции величин и статистической значимости их взаимосвязи 
# Делаем вывод на основе сравнения критического значения с расчетным
if abs(t_stat) > t_crit:
    print('На основании t-критического, присутствует статистическая значимость взаимосвязи x и y.')
    print()
else:
    print('На основании t-критического, отсутствует статистическая значимость взаимосвязи x и y.')
    print()


# Делаем вывод на основе сравнения p-value и уровня значимости
if p_value < alpha:
    print('На основании p-value при уровне значимости 0.05, присутствует статистическая значимость взаимосвязи x и y.')
    print()
else:
    print('На основании p-value при уровне значимости 0.05, отсутствует статистическая значимость взаимосвязи x и y.')
    print()

print("""Вычисленный коэффициент корреляции ~0.8, что свидетельствует о сильной корреляции между x и y""")
```

```python
# КРИТЕРИЙ 12
import statsmodels.api as sm

# расчет стандартных коэффицентов регрессии
# Создаем матрицу наблюдений (факторов) X и вектор правильных ответов (целевую переменную) y
X = df4[['x1', 'x2']]
y = df4['y']

# Стандартизируем переменные
X_std = (X - X.mean()) / X.std()
y_std = (y - y.mean()) / y.std()

# Создаём модель множественной регрессии
model = sm.OLS(y_std, X_std).fit()
b1 = float(model.params.iloc[0])
b2 = float(model.params.iloc[1])

print(f"""Стандартные коэффициенты регрессии для:
x1: {b1:.4f},
x2: {b2:.4f}""")
```

```python
# КРИТЕРИЙ 13

# расчет коэффициентов "чистой" регрессии
# Добавляем константу
X = sm.add_constant(X)

# Создаём модель множественной регрессии
model = sm.OLS(y, X).fit()
b0 = float(model.params.iloc[0])
b1 = float(model.params.iloc[1])
b2 = float(model.params.iloc[2])

print(f"""Стандартные коэффициенты регрессии для:
Константа: {b0:.4f}
x1: {b1:.4f},
x2: {b2:.4f}""")


# построение диаграмм рассеяния прогнозных и реальных значений на одном графике
y_pred = model.predict(X)
plt.scatter(y, y_pred, color='orange', label='Прогнозные и реальные значения')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'b--', label='Идеальная модель')
plt.title('Диаграмма рассеяния прогнозных и реальных значений')
plt.xlabel('Реальные значения')
plt.ylabel('Прогнозные значения')
plt.legend()
plt.show()
```

```python
# КРИТЕРИЙ 14
from sklearn.metrics import r2_score

# расчет индекса множественной корреляции
r2 = r2_score(y, y_pred)
R = r2**0.5

print(f'Индекс множественной корреляции R: {R:.4f}')

# расчет F-критерия
f_stat = model.fvalue
f_crit = f.ppf(0.95, 2, len(df4)-3)
print(f'f-критерий: {f_stat:.4f}')
print(f'f-критическое: {f_crit:.4f}')

# вывод о статистической значимости уравнения в целом
print("""На основании значений Индекса множественной корреялиции, а также F-критерия, можно сказать, 
что уравнение статистически значимо""")
```
