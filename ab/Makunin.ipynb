**Часть 1**  

Применение метода Bootstrap состоит из следующих шагов.

1. Для начала необходимо подготовить анализируемые данные, проверить что они не содержат аномальных значений или пропусков.
2. Сформулировать Н0 и Н1-гипотезы.
3. В каждой из анализируемых выборок выбрать множество подвыборок, для которых рассчитывается необходимая статистика.
4. Полученные результаты необходимо проанализировать, сравнить с Н0 и на основании данного сравнения сделать вывод о принятии/непринятии Н0.

**Часть 2**

* Какие преимущества дает использование метода Bootstrap?  
Bootstrap-тест позволяет просто и быстро оценивать самые разные статистики (доверительные интервалы, дисперсию, корреляцию) для сложных моделей, данных, имеющих сложную структуру или распределение. При проведении bootstrap-тестов отсутствует необходимость строить предположения о распределении данных, при этом bootstrap эффективен даже при анализе относительно небольших объемов данных.
В то же время, если анализируемый объем данных сравнительно велик, то применение bootstrap-метода потребует меньше вычислительных мощностей и займет меньше времени по сравнению с применением традиционных методов. Также, при большом количестве подвыборок и итераций, появляется эффект самокорреляции

* С какими метриками можно оперировать при использовании Bootstrap метода?  
При проведении bootstrap-анализа можно оперировать метриками, которые помогают оценить характер взаимодействия между данными в выборках.
Ключевыми из них являются: среднее, медиана, мода, коэффициент корреляции, p-value(критерий значимости).

* Какого размера должны быть bootstrap семплы?  
В зависимости от поставленной задачи, могут отличаться и размеры сэмплов. Обычно, это порядка нескольких сотен или тысяч элементов. В общем же случае размер выборки должен быть достаточным для надежной оценки статистических характеристик.

* Допустимо ли повторное использование одного и того же значения из исходной выборки в одном Bootstrap-семпле?  
Так как множество подвыборок чаще всего создаётся набором случайно выбранных значений, то повторное использование одного и того же значения вполне допустимо, если это не нарушает независимости выборки.
