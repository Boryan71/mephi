При биномиальном распределении вероятность можно вычислить следующим образом:  

<p style="text-align: center;">$P(X) = C_n^k \cdot p^k \cdot q^{n-k}$</p>

где:
* $C_n^k$ — число сочетаний из $n$ по $k$,  
* $p$ — вероятность выбора техники с подключением,  
* $q$ — вероятность выбора техники без подключения = $1−p$,  
* $n$ — общее количество испытаний,  
* $k$ — количество успехов.

* $P(X=0)$,

<p style="text-align: center;">$P(X = 0) = C_7^0 \cdot (\frac{1}{4})^0 \cdot (\frac{3}{4})^{7-0} = (\frac{3}{4})^7$</p>

Суммируя полученные значения, находим итоговую вероятность:

<p style="text-align: center;">$P(X \leq 2) = (\frac{3}{4})^7 + \frac{7}{4} \cdot (\frac{3}{4})^6 + \frac{21}{16} \cdot (\frac{3}{4})^5 = 0.756$</p>

**Вопрос 1**.

Для решения этой задачи подходит биномиальное распределение, так как мы имеем фиксированное количество независимых событий (каждый ноутбук либо ломается, либо нет), каждое из которых может иметь только два исхода (ноутбук сломался/не сломался).  
Вероятность поломки одного ноутбука в течение месяца равна $p=0.0002$, а количество ноутбуков равно $n=10000$.  
Воспользуемся формулой биномиальной вероятности:

<p style="text-align: center;">$P(X = k) = C_n^k \cdot p^k \cdot (1-p)^{n-k}$</p>

В нашем случае $k=1$ (вероятность, что сломается ровно 1 ноутбук):

<p style="text-align: center;">$P(X = 1) = C_{10000}^1 \cdot (0.0002)^1 \cdot (0.9998)^{10000-1}$</p>

```python
p = 0.0002
n = 10000
k = 1

P = stats.binom.pmf(k, n, p)

print(f'Вероятность, что сломается ровно 1 ноутбук: {P:.5f}')
```
